```python
import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import math as mt
```

```python
data=np.loadtxt('/raid2/ymao/VIC_RBM_east_RIPS/RIPS/model_run/output/vic/Maurer_0.125/Tennessee/fluxes_1949_2010/flow_no_header/flow_35.8125_-87.3125')
```

```python
n = len(data[:,0])
print n
```

```python
d1 = datetime.date(int(mt.floor(data[0,0])),int(mt.floor(data[0,1])),int(mt.floor(data[0,2])))
n1=mdates.date2num(d1)
print d1, n1
#date = []
#numdays = 10
#for d in 0:numdays
#    date.append(now - datetime.timedelta(days = d))
```

```python
date = []
for i in np.arange(n):
    day = datetime.date(int(mt.floor(data[i,0])),int(mt.floor(data[i,1])),int(mt.floor(data[i,2])))
    numday = mdates.date2num(day)
    date.append(numday)
```

```python
# plot precipitation
prep = data[:,3]
fig = plt.figure()
plt.xlabel('Time')
plt.ylabel('Prep(mm/day)')
plt.plot_date(date,prep,'-')
fig.autofmt_xdate()
plt.show()
```

```python
# plot E
evap = data[:,4]
fig = plt.figure()
plt.xlabel('Time')
plt.ylabel('evap(mm/day)')
plt.plot_date(date,evap,'-')
fig.autofmt_xdate()
plt.show()
```

```python
# plot Total Runoff
TR = data[:,5]+data[:,6]
fig = plt.figure()
plt.xlabel('Time')
plt.ylabel('TR(mm/day)')
plt.plot_date(date,TR,'-')
fig.autofmt_xdate()
plt.show()
```
